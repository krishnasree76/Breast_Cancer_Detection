```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
```

Loading the Dataset

```python
# Load the Dataset
file_path = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
column_names = ["ID", "Diagnosis"] + [f"Feature_{i}" for i in range(1, 31)]
data = pd.read_csv(file_path, header=None, names=column_names)
```

Outlier Detection

```python
# Calculate and visualize the impact of outliers
# Define IQR method for outlier detection
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
```

```python
# Identify outliers
outliers = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))
print("\nOutliers detected:")
print(outliers.sum())
```

Outlier Visualization

```python
# Get the number of outliers for each column
num_outliers_per_column = outliers.sum()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=num_outliers_per_column.index, y=num_outliers_per_column.values)
plt.xticks(rotation=90)
plt.title("Number of Outliers per Feature")
plt.show()
```

```python
# Split data into features and labels
X = data.drop('Diagnosis', axis=1)
y = data['Diagnosis']
```

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Display the first few rows of the scaled features
print(pd.DataFrame(X_scaled, columns=X.columns).head())
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Applying SMOTE on training data only
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Initialize AdaBoost classifier
ada_classifier = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),  # Decision stump as weak learner
    n_estimators=50,  # Number of weak learners
    learning_rate=1.0,  # Default learning rate
    algorithm="SAMME",  # Set to SAMME to avoid the FutureWarning
    random_state=42
)
```

```python
# Training the model with SMOTE-applied data
ada_classifier.fit(X_train_smote, y_train_smote)
```

```python
# Predicting on the test set
y_pred_smote = ada_classifier.predict(X_test)
```

```python
# Calculating performance metrics
accuracy_smote = accuracy_score(y_test, y_pred_smote)
precision_smote = precision_score(y_test, y_pred_smote)
recall_smote = recall_score(y_test, y_pred_smote)
f1_smote = f1_score(y_test, y_pred_smote)
conf_matrix_smote = confusion_matrix(y_test, y_pred_smote)
```

```python
# Displaying the evaluation results
print(f"\nAccuracy: {accuracy_smote:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix_smote)

# Visualize the Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_smote, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for AdaBoost Classifier')
plt.show()

print(f"\nPrecision: {precision_smote:.4f}")
print(f"\nRecall: {recall_smote:.4f}")
print(f"\nF1 Score: {f1_smote:.4f}")
```

Feature and Label Separation

```python
# Split data into features and labels
X = data_no_outliers.drop('Diagnosis', axis=1)
y = data_no_outliers['Diagnosis']
```

Feature Scaling

```python
# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Display the first few rows of the scaled features
print(pd.DataFrame(X_scaled, columns=X.columns).head())
```

Data Splitting

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

Applying SMOTE

```python
# Applying SMOTE on training data only
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

Initializing AdaBoost Classifier

```python
# Initialize AdaBoost classifier
ada_classifier = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),  # Decision stump as weak learner
    n_estimators=50,  # Number of weak learners
    learning_rate=1.0,  # Default learning rate
    algorithm="SAMME",  # Set to SAMME to avoid the FutureWarning
    random_state=42
)
```

Training the Model

```python
# Training the model with SMOTE-applied data
ada_classifier.fit(X_train_smote, y_train_smote)
```

Predicting on the Test Set

```python
# Predicting on the test set
y_pred = ada_classifier.predict(X_test)
```

Calculating Performance Metrics

```python
# Calculating performance metrics
accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

Displaying Evaluation Results

```python
print(f"\nAccuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(confusion)

# Visualize the Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for AdaBoost Classifier')
plt.show()

print(f"\nPrecision: {precision:.4f}")
print(f"\nRecall: {recall:.4f}")
print(f"\nF1 Score: {f1:.4f}")
```

Plotting ROC Curve

```python
# ROC Curve
# Get predicted probabilities
y_pred_proba = ada_classifier.predict_proba(X_test)[:, 1]  # Probability of the positive class
```

```python
# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
```

```python
# Calculate the AUC
roc_auc = auc(fpr, tpr)
```

```python
# Plot the ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

# Set up the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of weak learners (trees)
    'learning_rate': [0.1, 0.5, 1.0, 2.0],  # The learning rate to scale the contribution of each weak learner
    'estimator__max_depth': [1, 3, 5, 7]  # Max depth of the base decision tree classifier (weak learner)
}

# Initialize the DecisionTreeClassifier for the base estimator (decision stump)
base_estimator = DecisionTreeClassifier(max_depth=1, random_state=42)

# Initialize the AdaBoost classifier
ada_classifier = AdaBoostClassifier(
    estimator=base_estimator,  # Pass the base estimator here
    random_state=42
)

# Use GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(estimator=ada_classifier, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)

# Fit the GridSearchCV model
grid_search.fit(X_train_smote, y_train_smote)

# Get the best parameters found by GridSearchCV
print(f"Best Parameters: {grid_search.best_params_}")

# Train the AdaBoost model using the best parameters
best_ada_classifier = grid_search.best_estimator_

# Predict on the test set
y_pred_best = best_ada_classifier.predict(X_test)

# Calculate performance metrics
accuracy_best = accuracy_score(y_test, y_pred_best)
precision_best = precision_score(y_test, y_pred_best)
recall_best = recall_score(y_test, y_pred_best)
f1_best = f1_score(y_test, y_pred_best)
conf_matrix_best = confusion_matrix(y_test, y_pred_best)

# Display the evaluation results for the best model
print(f"\nAccuracy (Best Model): {accuracy_best:.4f}")
print("\nConfusion Matrix (Best Model):")
print(conf_matrix_best)

# Visualize the Confusion Matrix for the best model
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix_best, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Best AdaBoost Classifier')
plt.show()

print(f"\nPrecision (Best Model): {precision_best:.4f}")
print(f"\nRecall (Best Model): {recall_best:.4f}")
print(f"\nF1 Score (Best Model): {f1_best:.4f}")
```

```python
# Predicting on the test set
y_pred_test = grid_search.best_estimator_.predict(X_test)

# Accuracy on test data
accuracy_test = accuracy_score(y_test, y_pred_test)
print(f"Test Accuracy: {accuracy_test:.4f}")

# Confusion Matrix for the Test Set
cm_test = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix (Test Set):")
print(cm_test)

# Precision, Recall, F1 Score for the Test Set
precision_test = precision_score(y_test, y_pred_test)
recall_test = recall_score(y_test, y_pred_test)
f1_test = f1_score(y_test, y_pred_test)

print(f"Precision (Test Set): {precision_test:.4f}")
print(f"Recall (Test Set): {recall_test:.4f}")
print(f"F1 Score (Test Set): {f1_test:.4f}")
```

```python
from sklearn.metrics import classification_report

# Classification Report for the Test Set
report = classification_report(y_test, y_pred_test)
print("Classification Report (Test Set):")
print(report)
```
